```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
experiment_setting = ["C", "V", "CV", "VC"]
model_list = [f"MODEL{i}" for i in range(5)]
comparison_method = ["DAR", "mDAR", "DARC", "DARC_SDE", "FEW_SHOT_LEARNED", "MAML_SDE", "ONLY_TARGET"]
source_method = ["SOURCE", "mSOURCE"]
```

```python
total_result_df = pd.DataFrame()
for comparison_method_i in comparison_method:
    temp_df = pd.DataFrame()
    for experiment_setting_i in experiment_setting:
        temp_df = pd.concat([temp_df, pd.read_csv(f"{comparison_method_i}_{experiment_setting_i}_RESULT.csv")])
    temp_result_df = temp_df.describe().loc[['25%','50%','75%','mean','std']]
    temp_result_df.columns=[f"{comparison_method_i+' '+temp_df.columns[0]}",f"{comparison_method_i+' '+temp_df.columns[1]}"]
    total_result_df = pd.concat([total_result_df, temp_result_df], axis=1)
total_result_df = total_result_df.transpose()
total_result_df
```

```python
comparison_method = ["ONLY_TARGET", "FEW_SHOT_LEARNED", "DARC", "DARC_SDE", "MAML_SDE"]
total_result_df = pd.DataFrame()
for comparison_method_i in comparison_method:
    temp_df = pd.DataFrame()
    for experiment_setting_i in experiment_setting:
        temp_df = pd.concat([temp_df, pd.read_csv(f"{comparison_method_i}_{experiment_setting_i}_RESULT.csv")])
    temp_result_df = temp_df.copy().reset_index(drop=True)
    temp_result_df.columns=[f"{comparison_method_i+' '+temp_df.columns[0]}",f"{comparison_method_i+' '+temp_df.columns[1]}"]
    total_result_df = pd.concat([total_result_df, temp_result_df], axis=1)
comparison_method = ["ONLY", "SHOT", "DARC", "DARC*", "MAML*"]
plt.figure(figsize=(12,8))
plt.boxplot(np.array(total_result_df.transpose()).tolist())
plt.xticks([i for i in range(1,len(comparison_method)*2+1)], np.array([[i, i+"(d)"] for i in comparison_method]).reshape(-1))
plt.title('TARGET MSE')
plt.show()
comparison_method = ["DARC", "DARC_SDE", "FEW_SHOT_LEARNED", "MAML_SDE", "ONLY_TARGET"]
```

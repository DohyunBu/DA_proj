## MAML

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.h1 = nn.Linear(1, 40)
        self.relu = nn.ReLU()
        self.h2 = nn.Linear(40, 40)
        self.h3 = nn.Linear(40, 40)
        self.reg = nn.Linear(40, 1)
        
    def forward(self, x):
        x = self.h1(x)
        x = self.relu(x)
        x = self.h2(x)
        x = self.relu(x)
        x = self.h3(x)
        x = self.relu(x)
        x = self.reg(x)

        return x
```
